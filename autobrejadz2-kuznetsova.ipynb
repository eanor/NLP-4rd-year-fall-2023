```python
import whisper
from transformers import pipeline
import webvtt
import re
from jiwer import wer
import speech_recognition as sr
```

```python
r = sr.Recognizer()
with sr.AudioFile('Iconic British Moments #2-Mtm1Vl8SFzM.wav') as source:
    audio = r.record(source)
```

```python
# OpenAI Whisper library
whisper_audio = r.recognize_whisper(audio)
whisper_audio
```

```python
# Google Cloud Speech (with default key)
google_audio = r.recognize_google(audio)
google_audio
```

```python
#Hugging face
transcriber = pipeline("automatic-speech-recognition", "facebook/wav2vec2-base-960h")
hug_audio = transcriber('Iconic British Moments #2-Mtm1Vl8SFzM.wav')
print(hug_audio)
```

```python
auto_subs = []
for i in webvtt.read('Iconic British Moments #2-Mtm1Vl8SFzM.en.vtt'):
    auto_subs.append(i.text.split('\n'))
auto_subs
```

```python
subs = [item if item != ' ' else None for sublist in auto_subs for item in sublist]
```

```python
subs_list = [i for i in subs if i is not None]
```

```python
clean_subs = ' '.join(subs_list)
clean_subs 
```

```python
clean_subs = clean_subs.lower()
clean_subs = re.sub(r'[^\w\s]', '', clean_subs)
clean_subs = re.sub(r'  ', ' ', clean_subs)
clean_subs
```

```python
hug_audio = hug_audio['text'].lower()
hug_audio = re.sub(r'[^\w\s]', '', hug_audio)
hug_audio = re.sub(r'  ', ' ', hug_audio)

whisper_audio = whisper_audio.lower()
whisper_audio = re.sub(r'[^\w\s]', '', whisper_audio)
whisper_audio = re.sub(r'  ', ' ', whisper_audio)

google_audio = google_audio.lower()
google_audio = re.sub(r'[^\w\s]', '', google_audio)
google_audio = re.sub(r'  ', ' ', google_audio)
```
