```python
for word in my_corpus:
  for sent in corpus[word]:
    context = []
    context_lemmatized = []
    for token in sent.tokens:
      context.append(str(token.text))
      context_lemmatized.append(str(token.lemma))
    context = ' '.join(context)
    context_lemmatized = ' '.join(context_lemmatized)
    context = re.sub(r'\s+([!"#$%&\'()*+,-.\/:;<=>?@\[\\\]\^_`{\|}~])', r'', context)
    context_lemmatized = re.sub(r'\s+([!"#$%&\'()*+,-.\/:;<=>?@\[\\\]\^_`{\|}~])', r'', context_lemmatized)
    means = vm.disambiguate(word, context_lemmatized)
    i = np.argmax(means)
    df = pd.concat([pd.DataFrame([[word, context, context_lemmatized, i]], columns=df.columns), df], ignore_index=True)
```

```python
morph = MorphAnalyzer()
token = RegexpTokenizer('\w+')
stops = set(stopwords.words('russian'))

def normalize(text):
    words = [morph.parse(word)[0].normal_form for word in tokenize(text) if word]
    return words

def tokenize(text):
    return token.tokenize(text)
```

```python
model = ElmoModel()
model.load('196')
```

```python
def get_elmo_vectors(word, contexts, model):
    tokens = [normalize(c) for c in contexts]
    all_vectors = model.get_elmo_vectors(tokens)
    word_vecs = []
    for i in range(len(contexts)):
        try:
            word_vecs.append(all_vectors[i][tokens[i].index(word)])
        except ValueError:
            continue
    return word_vecs
```
